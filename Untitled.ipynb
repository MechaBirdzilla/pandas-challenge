```python
import pandas as pd
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# zip(student_data_df["math_score"], student_data_df["reading_score"])

passing_math = 0
passing_reading = 0
pass_both = 0
for (math_score, reading_score) in zip(student_data["math_score"], student_data["reading_score"]):
    if math_score >= 70:
        passing_math = passing_math + 1

    if reading_score >= 70:
        passing_reading = passing_reading + 1

    if reading_score >= 70 and math_score >= 70:
        pass_both = pass_both + 1

```

```python
district_summary = {
    "Total Schools": [school_data["school_name"].count()],
    "Total Students": [student_data["student_name"].count()],
    "Total Budget" : [school_data["budget"].sum()],
    "Average Math Score": [round(student_data["math_score"].mean(), 2)],
    "Average Reading Score": [round(student_data["reading_score"].mean(), 2)],
    "% Passing Math": [round(((passing_math / student_data["student_name"].count())*100), 2)],
    "% Passing Reading": [round(((passing_reading / student_data["student_name"].count())*100), 2)],
    "% Overall Passing": [round(((pass_both / student_data["student_name"].count())*100), 2)]
    }
district_summary
```

```python
df_district_summary = pd.DataFrame(data=district_summary)
df_district_summary
```

```python
#	Average Math Score	Average Reading Score	% Passing Math	% Passing Reading	% Overall Passing
test = school_data_complete.groupby(school_data_complete['school_name'])

school_summary = {
    "School Name": school_data["school_name"],
    "School Type": school_data["type"],
    "Total Students": school_data["size"],
    "Total Budget": school_data["budget"],
    "Budget per Student": school_data["budget"] / school_data["size"]
}
school_summary_df = pd.DataFrame(data=school_summary).set_index("School Name")
school_summary_df

student_summary = {
    "Average Math Score": round(test["math_score"].mean(),2),
    "Average Reading Score": round(test["reading_score"].mean(),2),
    "% Passing Math": round(school_data_complete[school_data_complete["math_score"] >= 70].groupby(school_data_complete['school_name'])["math_score"].count() / test["math_score"].count()*100, 2),
    "% Passing Reading": round(school_data_complete[school_data_complete["reading_score"] >= 70].groupby(school_data_complete['school_name'])["reading_score"].count() / test["reading_score"].count()*100, 2),
    "% Overall Passing": round(school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].groupby(school_data_complete['school_name'])["reading_score"].count() / test["reading_score"].count()*100, 2)

}

student_summary_df = pd.DataFrame(data=student_summary)
summary_df = school_summary_df.join(student_summary_df)
summary_df
```

```python
grades_dict = {"School Name": school_data["school_name"].sort_values()}
for grade in school_data_complete["grade"].unique():
    grades_dict[grade] = []

    for row in school_data_complete[school_data_complete["grade"] == grade].groupby("school_name")["reading_score"].mean():
        grades_dict[grade].append(round(row,2))

grades_df = pd.DataFrame(data=grades_dict).set_index("School Name")
grades_df
```

```python
grades_dict = {"School Name": school_data["school_name"].sort_values()}
for grade in school_data_complete["grade"].unique():
    grades_dict[grade] = []

    for row in school_data_complete[school_data_complete["grade"] == grade].groupby("school_name")["math_score"].mean():
        grades_dict[grade].append(round(row,2))
        
grades_df = pd.DataFrame(data=grades_dict).set_index("School Name")
grades_df
```

```python
school_spending_df = summary_df.copy().rename(columns={"Per Student Budget": "Spending Ranges (Per Student)"})

school_spending_df['Spending Ranges (Per Student)'] = pd.cut(school_spending_df["Budget per Student"], [0, 585, 630, 645, 680], labels = ["<$585", "$585-630", "$630-645", "$645-680"])

school_spending_grouped = school_spending_df.groupby(["Spending Ranges (Per Student)"])

cbfkrjwfe = {
    'Average Math Score': school_spending_grouped["Average Math Score"].mean(),
    'Average Reading Score': school_spending_grouped["Average Reading Score"].mean(),
    '% Passing Reading': school_spending_grouped["% Passing Reading"].mean(),
    '% Passing Math': school_spending_grouped["% Passing Math"].mean(),
    '% Overall Passing': school_spending_grouped["% Overall Passing"].mean()
}

pd.DataFrame(data=cbfkrjwfe)
```

```python
school_size_df = summary_df.copy().rename(columns={"Total Students": "School Size"})

school_size_df['School Size'] = pd.cut(school_size_df["School Size"], [0, 1000, 2000, 5000], labels = ["<1000", "1000-2000", "2000-5000"])

school_size_grouped = school_size_df.groupby(["School Size"])

cbfkrjwfe = {
    'Average Math Score': school_size_grouped["Average Math Score"].mean(),
    'Average Reading Score': school_size_grouped["Average Reading Score"].mean(),
    '% Passing Reading': school_size_grouped["% Passing Reading"].mean(),
    '% Passing Math': school_size_grouped["% Passing Math"].mean(),
    '% Overall Passing': school_size_grouped["% Overall Passing"].mean()
}

pd.DataFrame(data=cbfkrjwfe)
```
